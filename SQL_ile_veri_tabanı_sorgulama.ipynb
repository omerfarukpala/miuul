#  VERİ TİPLERİ

**********************
tam sayı veri tipleri
**********************

bigint : 8 Byte
- Min : -2^63
- Max :  2^63

int    : 4 Byte
- Min : -2^31 = -2,147,483,648
- Max :   2^31 =  2,147,483,647

smallint : 2 Byte
- -2^15 = -32,768
-  2^15 =   32,767

tinyint : 1 byte
- Min : 0
- Max : 255


bit : eğer tabloda 8 yada daha az bir kolonu varsa 1 byte, 8 den fazla ise 2 byte yer kaplar.

- 0 ya da 1 değerini alır.

**********************
string veri tipleri
**********************

char : 0 - 8000 arasında

- Tanımlandığı kadar Byte
- char(10 -> 10 Byte)

varchar : 0 - 8000 arasında
- girilen değer uzunluğu + 2 Byte

varchar(MAX) : 0 - 2 147 483 647 arasında
- Max değeri 2^31-1 Byte

text : 0 - 2 147 483 647 arasında
- Max değeri 2^31-1 Byte

ntext : 0 ile 1 073 741 823 arasında
- Max değeri 2^30-1 Byte

** text ve ntext artık kullanılmıyor.

# Kategorik Değişken Görselleştirme

```python
pd.set_option('display.max_columns', None)
# sağ tarafa doğru daha fazla değişken göremek için yapıyoruz
pd.set_option('display.width', 500)
df = sns.load_dataset("titanic")
df.head()
```

```python
df['sex'].value_counts()
df['sex'].value_counts().plot(kind='bar')
plt.show()
```

# Sayısal Değişken Görselleştirme

sayisal degisknlerde veri goresllestirme ile elde edilecek olan sey, degiskenin dagilimi ve yapisidir.

degiskenin dagilimini kolayca veren metotlar hist() ve boxplot() dur.

boxplot() ayrica aykiri degerleri de gosterir

matplotlib ile gorsellestirme

histogram

```python
# kütüphaneler
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
df = sns.load_dataset("titanic")
df.head()
```

```python
plt.hist(df["age"])
plt.show()
```

```python
plt.hist(df['age'], bins=20)  # sayisal degiskenlerin dagilimi elde edilir
plt.show(block=True)
sns.histplot(df['age'])
plt.show(block=True)
```

kutu grafik (aykırı değerleri görelleştiryor)

```python
# kutu grafik (aykırı değerleri görelleştiryor)
plt.boxplot(df["fare"])
plt.show()
```

# subplot

birden fazla görselin birlikte gösterilmesi

```python
# plot 1
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 2, 1)
plt.title("1")
plt.plot(x, y)

# plot 2
x = np.array([8, 8, 9, 9, 10, 15, 11, 15, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30, 30])
plt.subplot(1, 2, 2)
plt.title("2")
plt.plot(x, y)
plt.show()
```

3 grafiği bir satır 3 sütun olarak konumlamak.

```python
# 3 grafiği bir satır 3 sütun olarak konumlamak.
# plot 1
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 3, 1)
plt.title("1")
plt.plot(x, y)

# plot 2
x = np.array([8, 8, 9, 9, 10, 15, 11, 15, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30, 30])
plt.subplot(1, 3, 2)
plt.title("2")
plt.plot(x, y)

# plot 3
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 3, 3)
plt.title("3")
plt.plot(x, y)

plt.show()
```

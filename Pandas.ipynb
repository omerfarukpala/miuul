# iloc & loc

iloc : databaselerde index bilgisi vererek seçim yapmak için

loc : label based selection

```python
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
df = sns.load_dataset("titanic")
df.head()
```

```python
# iloc: integer based selection
df.iloc[0:3]
```

```python
df.iloc[0, 0]
```

```python
# loc: label based selection
df.loc[0:3]
```

```python
# df.iloc[0:3, "age"]         # hata alırız.  bizden bir index bilgisi istiyor.
df.iloc[0:3, 0:3]             # index bilgisi verdiğimde
```

```python
df.loc[0:3, "age"]          # label based yaptık.
```

```python
col_names = ["age", "embarked", "alive"]
df.loc[0:3, col_names]
```

# Koşullu Seçim (Conditional Selection)

```python
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
df = sns.load_dataset("titanic")
df.head()
```

```python
df[df["age"] > 50].head()          # yaşı 50 den büyük olan kişileri seçtik
df[df["age"] > 50]["age"].count()  # kaç kişi var?
```

```python
df.loc[df["age"] > 50, ["age", "class"]].head()   # bir koşul 2 sütunseçtik
```

* Birden fazla koşul giriyorsan bu koşullar parantez içine almak gerekir

```python
df.loc[(df["age"] > 50) & (df["sex"] == "male"), ["age", "class"]].head()   # yaşı 50 den büyük ve erkek koşulu
# birden fazla koşul giriyorsan bu koşullar parantez içine almak gerekir
```

```python
df["embark_town"].value_counts()
```

```python
df_new = df.loc[(df["age"] > 50) & (df["sex"] == "male")
       & ((df["embark_town"] == "Cherbourg") | (df["embark_town"] == "Southampton")),
       ["age", "class", "embark_town"]]
df_new
```

```python
df_new["embark_town"].value_counts()
```

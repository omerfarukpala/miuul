# Toplulaştırma & Gruplama (Aggregation & Groupby)

```python
df = sns.load_dataset("titanic")
```

 değişkenin ortalaması

```python
df["age"].mean()
```

cinsiyet değişkeninin yaşa göre ortalamaları.

```python
df.groupby("sex")["age"].mean()
```

```python
df.groupby("sex").agg({"age":"mean"})
```

 cinsiyete göre yaşların ortalamasını ve toplamını al

```python
df.groupby("sex").agg({"age":["mean", "sum"]})
```

cinsiyete göre yaş ortalamasını, toplamını al ve hayatta kalma oranlarının ortalamasını al

```python
df.groupby("sex").agg({"age":["mean", "sum"],
                       "survived": "mean"})
```

Cinsiyete ve şehrine göre yaş ve hayatta kalma ortalamaları

```python
df.groupby(["sex", "embark_town"]).agg({"age":["mean", "sum"],
                       "survived": "mean"})
```

Cinsiyete şehrine ve sınıfına göre yaş ve hayatta kalma ortalamaları

```python
df.groupby(["sex", "embark_town", "class"]).agg({"age":["mean", "sum"],
                       "survived": "mean"})
```

belirtilen şehir ve sınıfta bulunan kişi sayısı

ilk satırda 43 kişinin %97 si kurtulmuş.

Erkeklerde 265 kişinin sadece %12 si kurtulmuş

"Garibanın yüzü gülür mü" - Vahit Keskin

```python
df.groupby(["sex", "embark_town", "class"]).agg({
    "age":["mean", "sum"],
    "survived": "mean",
    "sex":"count"})
```

#Cut & QCut

Cut ve qcut fonksiyonları elimizdeki sayısal değişkenleri kategorik değişkenlere çevirmek için kullanılan en yaygın 2 ayrı fonksiyondur.

# Apply & Lambda

```python
df = sns.load_dataset("titanic")
```

```python
df["age2"] = df["age"] * 2
df["age3"] = df["age"] * 5

df.head()
```

parantez içine alarak sonuçları görebiliriz

```python
(df["age"] / 10).head()
(df["age2"] / 10).head()
(df["age3"] / 10).head()
```

```python
for col in df.columns:
  if "age" in col:
    print(col)
```

```python
for col in df.columns:
  if "age" in col:
    print((df[col] / 10).head())
```

```python
for col in df.columns:
  if "age" in col:
    df[col] = df[col] / 10
df.head()
```

Seçilen sütunlara lambdada verilen fonksiyon apply ile uygula.

```python
df[["age", "age2", "age3"]].apply(lambda x: x/10).head()
```

Age içeren değişkenlere lambdada verilen fonksiyonu apply ile uygula.

```python
df.loc[:, df.columns.str.contains("age")].apply(lambda x: x / 10).head()
```

Standartlaştırma fonksiyonu

```python
df.loc[:, df.columns.str.contains("age")].apply(lambda x: (x - x.mean()) / x.std()).head()
```

```python
def standart_scaler(col_name):
  return (col_name - col_name.mean()) / col_name.std()
```

```python
df.loc[:10, df.columns.str.contains("age")].apply(standart_scaler)
```

```python
df.loc[:, ["age", "age2", "age3"]] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head()
```

```python
df.loc[:, df.columns.str.contains("age")] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head()

df.head()
```

# merge

```python
df1 = pd.DataFrame({"employees": ["john", "dennis", "mark", "maria"],
                    "group": ["accounting", "engineering", "engineering", "hr"]})
```

```python
df2 = pd.DataFrame({"employees": ["mark", "john", "dennis", "maria"],
                    "start_date": [2010, 2009, 2014, 2019]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

on -> 2 dataframe dede ortak olan değişkene göre birleştir demektir.

 Burada çalışanlara göre birleştir demiş oluyoruz.

```python
pd.merge(df1, df2, on = "employees")
```

Amaç: Her çalışanın müdürünün bilgisine erişmek istiyoruz.

```python
df3 = pd.merge(df1, df2)
```

```python
df4 = pd.DataFrame({"group": ["accounting", "engineering", "hr"],
                    "manager": ["Caner", "Mustafa", "Berkcan"]})
```

```python
pd.merge(df3, df4)
```
